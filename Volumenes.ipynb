```python
# Importaci√≥n de librerias
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
plt.style.use('ggplot')
import pandas as pd
```

```python
# Lectura de datos almacenados en Github

datos_alcoa = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/ALCOA.csv', index_col=0) 
datos_alibaba = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/ALIBABA.csv', index_col=0) 
datos_amd = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/AMD.csv', index_col=0) 
datos_devon = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/DEVON.csv', index_col=0) 
datos_exxon = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/EXXON.csv', index_col=0) 
datos_facebook = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/FACEBOOK.csv', index_col=0) 
datos_intel = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/INTEL.csv', index_col=0) 
datos_netflix = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/NETFLIX.csv', index_col=0) 
datos_qualcomm = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/QUALCOMM.csv', index_col=0) 
datos_twitter = pd.read_csv('https://raw.githubusercontent.com/andresem/SemilleroTrading-2021-1/main/datos/TWITTER.csv', index_col=0) 

datos_globales =[datos_alcoa, datos_alibaba, datos_amd, datos_devon, datos_exxon,
                 datos_facebook, datos_intel, datos_netflix, datos_qualcomm, datos_twitter]
nombres = ['ALCOA', 'ALIBABA', 'AMD', 'DEVON', 'EXXON', 'FACEBOOK', 'INTEL', 'NETFLIX', 'QUALCOMM', 'TWITTER']

for datos in datos_globales:
  datos.index = pd.to_datetime(datos.index)

datos_globales[0].info()
```

```python
# # Diagrama de caja y bigotes del volumen por hora

for datos, nombre in zip(datos_globales, nombres):
  fig, axs = plt.subplots()
  datos['Hour'] = datos.index.hour.astype(str)
  datos.boxplot(column = ['Volume'], by = ['Hour'], ax = axs)
  axs.set_title(nombre)
  plt.show()
```

```python
# Distribuci√≥n del volumen por hora

for datos, nombre in zip(datos_globales, nombres):

  fig, axs = plt.subplots(3, 3)
  unicos = sorted(pd.unique(datos['Hour'].astype(int)))
  
  for i in range(len(unicos)):

    axs.reshape(-1)[i].hist(datos['Volume'][datos['Hour'] == str(unicos[i])], bins = 30)
    axs.reshape(-1)[i].set_title('Volume at ' + str(unicos[i]) + ' - Average volume: ' + 
                                str(round(datos['Volume'][datos['Hour'] == str(unicos[i])].mean())))
    axs.reshape(-1)[i].grid(True)

  fig.delaxes(axs[2,2])
  fig.delaxes(axs[2,1])
  fig.tight_layout(pad = 5)

  fig.suptitle('Distribuci√≥n del volumen por hora ' + nombre, fontsize=16)
  plt.show()
```
